# Интернет-магазин "Стримчик"

Интернет-магазин «Стримчик», продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (данные до 2016 года).<br> <br>  
**Цель исследования** - выявить определяющие успешность игры закономерности, которые позволят сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.<br> 
<br> 
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».<br> 
<br> 
**Описание данных**<br> 
<br> 
•	`Name` — название игры<br> 
•	`Platform` — платформа<br> 
•	`Year_of_Release` — год выпуска<br> 
•	`Genre` — жанр игры<br> 
•	`NA_sales` — продажи в Северной Америке (миллионы проданных копий)<br> 
•	`EU_sales` — продажи в Европе (миллионы проданных копий)<br> 
•	`JP_sales` — продажи в Японии (миллионы проданных копий)<br> 
•	`Other_sales` — продажи в других странах (миллионы проданных копий)<br> 
•	`Critic_Score` — оценка критиков (максимум 100)<br> 
•	`User_Score` — оценка пользователей (максимум 10)<br> 
•	`Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board).

##### 2.2.5 Столбец `rating`

Постмотрим на уникальные значения столбца `rating` и их количество в датасете.

**Entertainment Software Rating Board (ESRB)** — негосударственная организация, основное направление деятельности — принятие и определение рейтингов для компьютерных видеоигр и другого развлекательного программного обеспечения в США и Канаде. ESRB была основана в 1994 году ассоциацией Interactive Digital Software Association (современное название — «Entertainment Software Association»).<br>
«**EC**» («Early childhood») — «Для детей младшего возраста»<br>«**E**» («Everyone») — «Для всех»<br> «**E10+**» («Everyone 10 and older») — «Для всех от 10 лет и старше»<br> «**T**» («Teen») — «Подросткам» <br>«**M**» («Mature») — «Для взрослых» <br> «**AO**» («Adults Only 18+») — «Только для взрослых»<br> «**RP**» («Rating Pending») — «Рейтинг ожидается»<br> «**K-A**» - Для детей – для тех, кто уже научился говорить и держать джойстик, эти игры уже могут вызывать интерес [позже в 1998 году был переименован в Е – для всех возрастов]

Часть пропусков в столбце `rating`  можно объяснить тем, что рейтинг ESRB начали присваивать, начиная с 1994 года. 

```python
df['rating'].unique()
```

Заменим  пропуски и категорию 'E' на категорию - 'no_info', категорию 'K-A' переименуем в категорию 'E', опираясь на данные Википедии.

```python
df['rating'] = df['rating'].replace(np.nan, 'no_info')
df['rating'] = df['rating'].replace('K-A', 'E')
df['rating'] = df['rating'].replace('RP', 'no_info')
```

```python
df['rating'].value_counts()
```

Переведем данные столбца `rating` в категориальный тип

```python
df['rating'] = df['rating'].astype('category')
```

```python
def regional_pie_of_sales(act_df, name, region):
    (act_df.plot.pie(
    y=region,
    legend=False,
    pctdistance=0.7, 
    labeldistance=1.1, 
    autopct='%1.1f%%', 
    figsize=(12,6),title=name));
```

```python
def lollipop_plot(data, column, region, labels):
    """Функция построения Lollipop графиков:

    Для использования функции необходимо передать ей
    следующие аргументы:
    
    data - массив исходных данных;
    region, column - названия столбцов, относительно которых будет
        проводиться анализ - 'название_столбца';
    labels - массив названий категориальных данных - ['Заголовок','x_name','y_name'];

    """
    
    
    data = (act_df.pivot_table(index = column, values = region, aggfunc = 'sum').reset_index().
            sort_values(by = region).tail())
    
    my_range=range(1,len(data.index)+1)
    fig = plt.figure(figsize=(15, 6)) 
    plt.hlines(y=my_range, xmin=0, xmax=data[region], color='green')
    plt.scatter(data[region], my_range, color='green', alpha=1)
    plt.yticks(my_range, data[column])
    plt.title(labels[0])
    plt.xlabel(labels[1])
    plt.ylabel(labels[2])
    plt.grid(True)
    plt.show()
```

### 5. Проверка гипотез

#### 5.1 **Гипотеза № 1** : Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем нулевую и альтернативную гипотезы так:

**H0: cредние пользовательские рейтинги платформ Xbox One и PC равны**

**H1 : cредние пользовательские рейтинги платформ Xbox One и PC различаются (не равны)**

```python
xone_var = st.variation(act_df.query('platform == "XOne"')['user_score'])
pc_var = st.variation(act_df.query('platform == "PC"')['user_score'])
print('Равны ли дисперсии двух совокупностей?',(xone_var * 0.05 < abs(xone_var - pc_var)))
```

```python
pc = act_df[act_df['platform'] == 'PC'] 
xone = act_df[act_df['platform'] == 'XOne']

alpha = .05
results = st.ttest_ind(pc['user_score'], xone['user_score'], equal_var=False, nan_policy='omit')
print('значение p-value: ', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем Нулевую гипотезу')
else:
    print('Не получилось отвергнуть Нулевую гипотезу')
```

Полученное значение p-value говорит о том, что вероятность получить такое или большее различие случайно - близка к 31%. Можно с вероятностью 69% сделать вывод, что средние пользовательские рейтинги платформ Xbox One и PC различаются. 

#### 5.2 **Гипотеза № 2** : Средние пользовательские рейтинги жанров Action и Sports  разные.

Сформулируем нулевую и альтернативную гипотезы так:

**H0: средние пользовательские рейтинги жанров Action и Sports равны**

**H1 : средние пользовательские рейтинги жанров Action и Sports различаются (не равны)**

```python
action_var = st.variation(act_df.query('genre == "action"')['user_score'])
sports_var = st.variation(act_df.query('genre == "sports"')['user_score'])
print('Равны ли дисперсии двух совокупностей?',(((action_var * 0.05) < abs(action_var - sports_var))))
```

```python
action = act_df[act_df['genre'] == 'Action'] 
sports = act_df[act_df['genre'] == 'Sports']

alpha = .05
results1 = st.ttest_ind(action['user_score'], sports['user_score'], equal_var=False, nan_policy='omit')
print('значение p-value: ', results1.pvalue)

if (results1.pvalue < alpha):
    print('Отвергаем Нулевую гипотезу')
else:
    print('Не получилось отвергнуть Нулевую гипотезу')
```

Полученное значение p-value говорит о том, что вероятность получить такое или большее различие случайно - близка к 0%. Можно с вероятностью 99% сделать вывод, что средние пользовательские рейтинги жанров Action и Sports различаются. 

#### 5.3 Вывод

В Гипотезе № 1 не получилось отвергнуть нулевую гипотезу.  Нельзя утверждать, что средние пользовательские рейтинги платформ Xbox One и PC равны. В Гипотезе №2 отвергли нулевую гипотезу - средние пользовательские рейтинги жанров Action и Sports различаются.
